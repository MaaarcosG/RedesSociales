#### Apple

```
categories = []
full_text = []
retweet_count = []
created_at = []
geo = []

#Buscamos tweets de Apple
for tweet in tweepy.Cursor(api.search,
                           q="Apple",
                           tweet_mode="extended",
                           lang='en').items(10000):
  categories.append("Apple")
  full_text.append(tweet._json["full_text"])
  retweet_count.append(tweet._json["retweet_count"])
  created_at.append(tweet._json["created_at"])
  geo.append(tweet._json["geo"])

#Guardamos toda la data en un dataframe
dataframe = pd.DataFrame({'category': categories,
                          'full_text':full_text,
                          'retweet_count': retweet_count,
                          'created_at':created_at,
                          'geo':geo})

#Guardamos el dataframe en un csv de Drive
dataframe.to_csv('data.csv')
!cp data.csv "drive/My Drive/"

print("Se guardo correctamente Apple en CSV")
```
